Para este ejercicio vamos a realizar la curva de la funcion 
$$y=\sin x $$

Para mas informacion en [wikipedia](https://es.wikipedia.org/wiki/Seno_(trigonometr%C3%ADa)) y su grafica es


![graficar-seno.png](graficar-seno.png)

y ahora vamos a realiar su grafica con ayuda de python con el tutorial del siguiente video [tutorial](https://www.youtube.com/watch?v=jJOfDD9isiM&ab_channel=RodrigoDorantes-Gilardi)

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-np.pi,np.pi,0.1)   # inicio, final y secuencia 
y = np.sin(x)
plt.title('ejemplo de la funci√≥n seno')
plt.plot(x,y)
plt.legend(['sin(x)'])
plt.grid()
plt.show()
```
